## Starter Code to Import Libraries and Load the Weather and Coordinates Data

```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests
from pathlib import Path

# Import API key
from api_keys import geoapify_key
```

```python
# Load the CSV file created in Part 1 into a Pandas DataFrame
csv_file = Path("output_data/cities.csv")
city_data_df = pd.read_csv(csv_file)

# Display sample data
city_data_df.head()
```

---

### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
%%capture --no-display

# Configure the map plot
# YOUR CODE HERE
map_plot = city_data_df.hvplot.points(
    x='Lng',
    y='Lat',
    geo=True,
    tiles='OSM',  # Use OSM tiles
    frame_width=700,
    frame_height=500,
    size='Humidity',     # Use Humidity values for the size of points
    scale=0.8,
    color='City',        # Use City names for the color of points
    hover_cols=['City', 'Humidity'],  # Show City and Humidity in tooltips
)

# Display the map
map_plot
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Narrow down cities that fit criteria and drop any results with null values
# YOUR CODE HERE
criteria_df = city_data_df.loc[
    (city_data_df['Max Temp'] > 25) &  # Max Temperature greater than 25C
    (city_data_df['Max Temp'] < 30) &  # Max Temperature less than 30C
    (city_data_df['Humidity'] < 50) &  # Humidity less than 50%
    (city_data_df['Wind Speed'] < 10)  # Wind Speed less than 10 m/s
]

# Drop any rows with null values
# YOUR CODE HERE
criteria_df.dropna(inplace=True)

# Display sample data
# YOUR CODE HERE
criteria_df
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
# YOUR CODE HERE
hotel_df = criteria_df.copy()

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
# YOUR CODE HERE
hotel_df = hotel_df.assign(Hotel_Name="")

# Display sample data
# YOUR CODE HERE
hotel_df
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
# Set parameters to search for a hotel
radius = 10000  # 10,000 meters (10 km)
params = {
    "apiKey": geoapify_key,
    "filter": "category=accommodation",
}

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # Get latitude, longitude from the DataFrame
    lat = row["Lat"]
    lng = row["Lng"]
    
    # Add filter and bias parameters with the current city's latitude and longitude to the params dictionary
    params["filter"] = f"distance.to.LocationPoint({lng},{lat}) < {radius}"
    params["bias"] = f"proximity:{lng},{lat}"
    
    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"

    # Make an API request using the params dictionary
    response = requests.get(base_url, params=params)
    name_address = response.json()
    
    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel_Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel_Name"] = "No hotel found"
        
    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel_Name']}")

# Display sample data
hotel_df
```
